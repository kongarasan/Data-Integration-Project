```python
import jellyfish
import pandas as pd

# Mediated schema columns
mediated_cols = ['title', 'author', 'rating', 'review_count', 'price', 'publication_year']

# Columns of source datasets
amazon_cols = ['Unnamed: 0', 'price', 'ranks', 'title', 'no_of_reviews', 'ratings', 'author', 'cover_type', 'year', 'genre']
goodreads_cols = ['source_URL', 'Readers Choice Votes', 'Readers Choice Category', 'Title', 'Author', 'Total Avg Rating', 'Number of Ratings', 'Number of Reviews', 'Number of Pages', 'Edition', 'Book Description', 'First Published date', 'Kindle Version and Price', 'Kindle Price', 'About the Author']
kindle_cols = ['asin', 'title', 'author', 'soldBy', 'imgUrl', 'productURL', 'stars', 'reviews', 'price', 'isKindleUnlimited', 'category_id', 'isBestSeller', 'isEditorsPick', 'isGoodReadsChoice', 'publishedDate', 'category_name']

def jaro_winkler_matrix(mediated, source):
    sim_matrix = pd.DataFrame(index=mediated, columns=source)

    for m_col in mediated:
        for s_col in source:
            score = jellyfish.jaro_winkler_similarity(m_col.lower(), s_col.lower())
            sim_matrix.loc[m_col, s_col] = score
    return sim_matrix.astype(float)  # Convert to float for better display

# Compute similarity matrices
amazon_sim = jaro_winkler_matrix(mediated_cols, amazon_cols)
goodreads_sim = jaro_winkler_matrix(mediated_cols, goodreads_cols)
kindle_sim = jaro_winkler_matrix(mediated_cols, kindle_cols)

# Display the matrices nicely in Jupyter Notebook
from IPython.display import display

print("Jaro-Winkler similarity with Amazon dataset:")
display(amazon_sim)

print("\nJaro-Winkler similarity with Goodreads dataset:")
display(goodreads_sim)

print("\nJaro-Winkler similarity with Kindle dataset:")
display(kindle_sim)
```

```python
def semantic_similarity_matrix(mediated, source):
    sim_matrix = pd.DataFrame(index=mediated, columns=source, dtype=float)

    # Pre-process columns into spaCy docs once
    mediated_docs = {col: nlp(col.replace('_', ' ')) for col in mediated}
    source_docs = {col: nlp(col.replace('_', ' ')) for col in source}

    for m_col, m_doc in mediated_docs.items():
        for s_col, s_doc in source_docs.items():
            sim = m_doc.similarity(s_doc)
            sim_matrix.loc[m_col, s_col] = sim
    return sim_matrix

# Compute semantic similarity matrices
amazon_sem_sim = semantic_similarity_matrix(mediated_cols, amazon_cols)
goodreads_sem_sim = semantic_similarity_matrix(mediated_cols, goodreads_cols)
kindle_sem_sim = semantic_similarity_matrix(mediated_cols, kindle_cols)

print("Semantic similarity with Amazon dataset:")
display(amazon_sem_sim)

print("\nSemantic similarity with Goodreads dataset:")
display(goodreads_sem_sim)

print("\nSemantic similarity with Kindle dataset:")
display(kindle_sem_sim)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def tfidf_cosine_similarity_matrix(mediated, source):
    # Combine all columns (mediated + source) for fitting the vectorizer
    all_cols = mediated + source

    # Vectorize column names using TF-IDF
    vectorizer = TfidfVectorizer(analyzer='word', token_pattern=r'\b\w+\b')
    tfidf_matrix = vectorizer.fit_transform(all_cols)

    # Split vectors into mediated and source parts
    mediated_vecs = tfidf_matrix[:len(mediated)]
    source_vecs = tfidf_matrix[len(mediated):]

    # Compute cosine similarity matrix (mediated x source)
    sim_matrix = cosine_similarity(mediated_vecs, source_vecs)

    # Convert to DataFrame for nice display
    sim_df = pd.DataFrame(sim_matrix, index=mediated, columns=source)
    return sim_df

# Compute similarity matrices
amazon_tfidf_sim = tfidf_cosine_similarity_matrix(mediated_cols, amazon_cols)
goodreads_tfidf_sim = tfidf_cosine_similarity_matrix(mediated_cols, goodreads_cols)
kindle_tfidf_sim = tfidf_cosine_similarity_matrix(mediated_cols, kindle_cols)

print("TF-IDF Cosine similarity with Amazon dataset:")
display(amazon_tfidf_sim)

print("\nTF-IDF Cosine similarity with Goodreads dataset:")
display(goodreads_tfidf_sim)

print("\nTF-IDF Cosine similarity with Kindle dataset:")
display(kindle_tfidf_sim)
```

```python
#Combining similarity matrices using MIN, MAX, AVG and WEIGHTER AVERAGE
```

```python
import numpy as np

def generate_combiners(sim_matrices, weights=None):

    index = sim_matrices[0].index
    columns = sim_matrices[0].columns
    matrix_stack = np.array([df.values for df in sim_matrices])

    combined = {
        'min': pd.DataFrame(np.min(matrix_stack, axis=0), index=index, columns=columns),
        'max': pd.DataFrame(np.max(matrix_stack, axis=0), index=index, columns=columns),
        'avg': pd.DataFrame(np.mean(matrix_stack, axis=0), index=index, columns=columns)
    }

    if weights:
        weighted = np.average(matrix_stack, axis=0, weights=np.array(weights))
        combined['weighted_avg'] = pd.DataFrame(weighted, index=index, columns=columns)

    return combined
```

```python
from IPython.display import display
import numpy as np

# Make sure you already ran this earlier:
# generate_combiners() function

# Define matcher results for each dataset
amazon_matchers = [amazon_sim, amazon_sem_sim, amazon_tfidf_sim]
goodreads_matchers = [goodreads_sim, goodreads_sem_sim, goodreads_tfidf_sim]
kindle_matchers = [kindle_sim, kindle_sem_sim, kindle_tfidf_sim]

# Define weights: Jaro (0.4), Semantic (0.3), TF-IDF (0.3)
weights = [0.4, 0.3, 0.3]

# Generate combined matrices
amazon_combined = generate_combiners(amazon_matchers, weights)
goodreads_combined = generate_combiners(goodreads_matchers, weights)
kindle_combined = generate_combiners(kindle_matchers, weights)
```
